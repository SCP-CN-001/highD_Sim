# Dragon Lake Parking Trajectory Data Analysis

## Basic Information

Dragon Lake Parking (DLP) dataset is collected  by University of California, Berkeley, CA, USA. "The entire DLP dataset contains 30 scenes, 317, 873 frames, 5, 188 agents, and 15, 383, 737 instances".

This repository provides an official and authorized version of a data sample from the DLP dataset. It also provides a visualized insight into the complete dataset. For those interested in accessing the complete dataset, please follow this [link](https://sites.google.com/berkeley.edu/dlp-dataset).


### Trajectory Types in DLP

- `Car`
- `Medium Vehicle`
- `Bus`
- `Motorcycle`
- `Bicycle`
- `Pedestrian`
- `Undefined`

```python
import sys

sys.path.append(".")

import warnings

warnings.filterwarnings("ignore")

import json

import matplotlib.transforms as mtransforms

from utils import dlp

with open("../map/map.config", "r") as f:
    configs = json.load(f)

data_path = "../trajectory/DLP"
img_path = "../img/DLP/DLP.jpeg"
trajectory_class = [
    "Car",
    "Medium Vehicle",
    "Bus",
    "Motorcycle",
    "Bicycle",
    "Pedestrian",
    "Undefined",
]
map_range = [-20, 150, -40, 90]
```

## Data Analysis

The data analysis will first present visualizations of both the map and trajectory data and then analysis the speed distribution and lane changing frequency of vehicles, categorized by their types.

```python
transform = mtransforms.Affine2D().scale(0.47).translate(-1, -1.9)
dlp.plot_map_and_trajectories(data_path, img_path, transform, trajectory_class, configs)
```

### Proportion of Trajectory Types

In the DLP dataset, the proportion of pedestrians is notably high, because many individuals park their cars in the lot and then walk to their destinations. The proportion of pedestrians exhibits significant fluctuations across different track files.files.

```python
dlp.plot_class_proportion(data_path, trajectory_class, configs)
```

### Speed Distribution

The speed of vehicles within the parking lot tends to be low due to the presence of crowded pedestrians and the narrow lanes. Notably, you can observe acceleration on the straight lanes, and conversely, a deceleration occurs at the intersections.

```python
dlp.plot_speed_distribution(map_range, data_path, "Car", configs)
```

```python
dlp.plot_speed_distribution(map_range, data_path, "Pedestrian", configs)
```

### Parking Denstity

The figure below illustrates the frequency of parking position occupancy. Evidently, positions closer to the entrance are favored and more frequently occupied.

```python
dlp.plot_parking_density(map_range, data_path, configs)
```
