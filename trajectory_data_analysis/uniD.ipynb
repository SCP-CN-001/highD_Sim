# UniD Trajectory Data Analysis

## Brief Introduction to UniD

UniD is "a new dataset of naturalistic road user trajectories recorded at the campus of the RWTH Aachen University".

This repository provides a data sample including less than 5% of the original dataset. It also provides a visualized insight into the dataset. For those interested in accessing the complete dataset, please follow this [link](https://www.unid-dataset.com/).

The corresponding paper of this paper is not released yet. 

### Trajectory Types in InD

In uniD, trajectories falls into these following categories:

- `bicycle`
- `bus`
- `car`
- `motorcycle`
- `pedestrian`
- `trailer`
- `truck`
- `van`

### Match Trajectory Data with Maps

All the trajectory files are collected in the same location, which is denoted as uniD_0.

```python
import sys

sys.path.append(".")

import warnings

warnings.filterwarnings("ignore")

import json

import matplotlib.transforms as mtransforms

from utils import levelx

with open("../map/map.config", "r") as f:
    configs = json.load(f)

data_path = "../trajectory/uniD/data"
img_path = "../img/uniD"
trajectory_types = [
    "car",
    "truck",
    "trailer",
    "bus",
    "van",
    "motorcycle",
    "bicycle",
    "pedestrian",
]
```

```python
transform = mtransforms.Affine2D().scale(0.2).translate(1, -143)
levelx.plot_map_and_trajectories(
    "uniD_0", data_path, img_path, transform, trajectory_types, configs
)
```

### Proportion of Trajectory Types

UniD stands out as an exceptional dataset within the LevelX series. Unlike the other datasets in the series, UniD contains a higher proportion of pedestrians compared to cars. This distinction primarily arises from the dataset's collection within a university scenario.

```python
levelx.plot_class_proportion("uniD_0", data_path, trajectory_types, configs)
```
